<font color = green> Importando as Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<font color = green> -Definindo os dados como um dicionário, onde cada chave representa uma coluna do DataFrame.

```python
data = {
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}
```

<font color = green> -Criando o DataFrame 'df' a partir do dicionário data.

```python
df = pd.DataFrame(data)
```

<font color = green> -Neste bloco, as bibliotecas necessárias são importadas: Pandas para manipulação de dados, NumPy para cálculos numéricos, Matplotlib para visualizações gráficas e Seaborn para aprimorar as visualizações. Em seguida, são definidos os dados em um dicionário chamado 'data', onde cada chave representa uma coluna do DataFrame. O DataFrame 'df' é criado a partir desses dados.

```python
# Cálculo do total de horas trabalhadas
total_horas_trabalhadas = df['Horas Trabalhadas'].sum()

# Cálculo da média diária de horas trabalhadas
media_horas_trabalhadas_diarias = df['Horas Trabalhadas'].mean()

# Cálculo do total de bugs corrigidos
total_bugs_corrigidos = df['Bugs Corrigidos'].sum()

# Cálculo da média diária de bugs corrigidos
media_bugs_corrigidos_diarios = df['Bugs Corrigidos'].mean()

# Cálculo do total de tarefas concluídas
total_tarefas_concluidas = df['Tarefas Concluídas'].sum()

# Cálculo da média diária de tarefas concluídas
media_tarefas_concluidas_diarias = df['Tarefas Concluídas'].mean()

# Cálculo da produtividade diária (Tarefas Concluídas por Hora)
df['Produtividade Diária'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']
```

<font color = green> -Neste bloco, são realizados cálculos para as principais métricas do projeto:

<font color = green> -total_horas_trabalhadas: calcula o total de horas trabalhadas somando os valores da coluna 'Horas Trabalhadas'.
media_horas_trabalhadas_diarias: calcula a média diária de horas trabalhadas obtendo a média da coluna 'Horas Trabalhadas'.
total_bugs_corrigidos: calcula o total de bugs corrigidos somando os valores da coluna 'Bugs Corrigidos'.
media_bugs_corrigidos_diarios: calcula a média diária de bugs corrigidos obtendo a média da coluna 'Bugs Corrigidos'.
total_tarefas_concluidas: calcula o total de tarefas concluídas somando os valores da coluna 'Tarefas Concluídas'.
media_tarefas_concluidas_diarias: calcula a média diária de tarefas concluídas obtendo a média da coluna 'Tarefas Concluídas'.
df['Produtividade Diária']: cria uma nova coluna chamada 'Produtividade Diária', que representa a relação entre tarefas concluídas e horas trabalhadas para cada dia.

```python
# Gráfico de barras para mostrar o total de horas trabalhadas por dia
plt.figure(figsize=(7, 6))
sns.barplot(x='Dia', y='Horas Trabalhadas', data=df)
plt.title('Total de Horas Trabalhadas por Dia')
plt.xlabel('Dia')
plt.ylabel('Total de Horas Trabalhadas')
plt.show()

# Gráfico de barras para mostrar o total de bugs corrigidos por dia
plt.figure(figsize=(7, 6))
sns.barplot(x='Dia', y='Bugs Corrigidos', data=df)
plt.title('Total de Bugs Corrigidos por Dia')
plt.xlabel('Dia')
plt.ylabel('Total de Bugs Corrigidos')
plt.yticks(np.arange(0, df['Bugs Corrigidos'].max() + 1, 1))
plt.show()

# Gráfico de barras para mostrar o total de tarefas concluídas por dia
plt.figure(figsize=(7, 6))
sns.barplot(x='Dia', y='Tarefas Concluídas', data=df)
plt.title('Total de Tarefas Concluídas por Dia')
plt.xlabel('Dia')
plt.ylabel('Total de Tarefas Concluídas')
plt.show()
```

<font color = green> -Neste bloco, três gráficos de barras são criados usando o Seaborn e o Matplotlib. Cada gráfico representa o total de horas trabalhadas, total de bugs corrigidos e total de tarefas concluídas por dia da semana.

```python
# Gráfico de barras para mostrar a média diária de horas trabalhadas
plt.figure(figsize=(4, 4))
sns.barplot(x=['Média'], y=[media_horas_trabalhadas_diarias])
plt.title('Média Diária de Horas Trabalhadas')
plt.ylabel('Horas')
plt.show()

# Gráfico de barras para mostrar a média diária de bugs corrigidos
plt.figure(figsize=(4, 4))
sns.barplot(x=['Média'], y=[media_bugs_corrigidos_diarios])
plt.title('Média Diária de Bugs Corrigidos')
plt.ylabel('Bugs Corrigidos')
plt.show()

# Gráfico de barras para mostrar a média diária de tarefas concluídas
plt.figure(figsize=(4, 4))
sns.barplot(x=['Média'], y=[media_tarefas_concluidas_diarias])
plt.title('Média Diária de Tarefas Concluídas')
plt.ylabel('Tarefas Concluídas')
plt.show()
```

<font color = green> -Neste bloco, são criados mais três gráficos de barras usando o Seaborn e o Matplotlib. Cada gráfico representa a média diária de horas trabalhadas, média diária de bugs corrigidos e média diária de tarefas concluídas.

```python
# Gráfico de barras para mostrar a produtividade diária (Tarefas Concluídas por Hora)
plt.figure(figsize=(7, 6))
sns.barplot(x='Dia', y='Produtividade Diária', data=df)
plt.title('Produtividade Diária (Tarefas Concluídas por Hora)')
plt.xlabel('Dia')
plt.ylabel('Tarefas Concluídas por Hora')
plt.show()

# Criação do relatório de progresso diário
relatorio = f"""
# Relatório de Progresso Diário

## Métricas

- Total de Horas Trabalhadas: {total_horas_trabalhadas} horas
- Média Diária de Horas Trabalhadas: {media_horas_trabalhadas_diarias:.2f} horas
- Total de Bugs Corrigidos: {total_bugs_corrigidos}
- Média Diária de Bugs Corrigidos: {media_bugs_corrigidos_diarios:.2f}
- Total de Tarefas Concluídas: {total_tarefas_concluidas}
- Média Diária de Tarefas Concluídas: {media_tarefas_concluidas_diarias:.2f}
"""

# Exibir o relatório diretamente no Colab
print(relatorio)
```
